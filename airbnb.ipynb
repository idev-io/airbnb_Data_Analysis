```python
# Import Required Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read ( Load ) Datasets 
dataset_path = r'datasets/AB_NYC_2019.csv'
airbnb = pd.read_csv(dataset_path)
```

```python
# Dataset exploration
airbnb.head()
```

```python
# Dataset info
airbnb.info()
```

```python
# Dataset Basic Statistics
airbnb.describe()
```

```python
# Check Dataset Null Values
airbnb.isna().sum()
```

```python
# Count values of specific columns
airbnb['neighbourhood_group'].value_counts()
```

```python
# Mean of `reviews_per_month` column
reviews_per_month_mean = airbnb['reviews_per_month'].mean()
reviews_per_month_mean
```

```python
# Fill `reviews_per_month` null values with mean
airbnb['reviews_per_month'].fillna(reviews_per_month_mean, inplace = True) 
```

```python
# Mode of `host_name` column
host_name_mode = airbnb['host_name'].mode()
host_name_mode
```

```python
# Fill `host_name` null values with mode
airbnb['host_name'].fillna(host_name_mode.values[0], inplace = True)
```

```python
# Mode of `name` column
name_mode = airbnb['name'].mode()
name_mode
```

```python
# Fill `name` null values with mode
airbnb['name'].fillna(name_mode.values[0], inplace = True)
```

```python
# Mode of `last_review` column
last_review_mode = airbnb['last_review'].mode()
last_review_mode
```

```python
# Fill `last_review` null vales with mode
airbnb['last_review'].fillna(last_review_mode.values[0], inplace = True)
```

```python
# Convert `last_review` column to datetime type 
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])
```

```python
# Export dataset after cleaning
airbnb.to_csv('datasets_after_cleaning/NYC_airbnb_after_cleaning.csv')
```

```python
grp_neighbourhood_group = airbnb.groupby('neighbourhood_group')['availability_365'].mean()
grp_neighbourhood_group
```

```python
# Get sum of `availability` when neighbourhood_group is Manhattan
airbnb.query('`neighbourhood_group` == "Manhattan"')['availability_365'].sum()
```

```python
# Data Visualization
# Bar plot of Mean of prices for each neighbourhood group

plt.figure(figsize = (15, 8))
sns.barplot(x="neighbourhood_group", y="price", data=airbnb, ci = False)

plt.title('Mean of prices for each neighbourhood group', color = '#341f97', fontsize=20)
plt.xlabel('Neighbourhood Groups', color = '#222f3e', fontsize=16)
plt.ylabel('Prices', color = '#222f3e', fontsize=16)

plt.show()
```

```python
# Bar plot of Mean of Number of Reviews for each neighbourhood group

plt.figure(figsize = (15, 8))
sns.barplot(x="neighbourhood_group", y="number_of_reviews", data=airbnb, ci = False)

plt.title('Mean of Number of Reviews for each neighbourhood group', color = '#341f97', fontsize=20)
plt.xlabel('Neighbourhood Groups', color = '#222f3e', fontsize=16)
plt.ylabel('Number of Reviews', color = '#222f3e', fontsize=16)

plt.show()
```

```python
# Bar plot of Mean of Prices for each Room Type

plt.figure(figsize = (15, 8))
sns.barplot(x="room_type", y="price", data=airbnb, ci = False)

plt.title('Mean of Prices for each Room Type', color = '#341f97', fontsize=20)
plt.xlabel('Room Type', color = '#222f3e', fontsize=16)
plt.ylabel('Prices', color = '#222f3e', fontsize=16)

plt.show()
```

```python
# Scatter plot of Prices Over Room Availability
plt.figure(figsize = (15, 8))
sns.scatterplot(data = airbnb, x = 'availability_365', y = 'price', hue = 'price')

plt.title('Prices Over Room Availability', color = '#1c0a23', fontsize=20)
plt.xlabel('Room Availability', color = '#222f3e', fontsize=16)
plt.ylabel('Prices', color = '#222f3e', fontsize=16)

plt.show()
```
